## Start Stimulation Recordings

```python
# Load stimulation configs
stim_configs = load_configs(path_config="../config/stimulation_configs.yaml")[TEST]

# Randomize the order of the tests
tests = [test for test in stim_configs]
random.shuffle(tests)

print(f"Tests to be conducted: {tests}")
```

```python
# Initialize recorder and start recording
for test_params in tests:
    print(f"Starting experiment {test_params}...")
    recorder_stim = StimulationActivityRecorder(electrode_selections_path=ELECTRODE_SELECTIONS_PATH,
                                                output_path=OUTPUT_PATH,
                                                chip_id=CHIP_ID,
                                                suffix=SUFFIX,
                                                div=DIV,
                                                stim_voltage=stim_configs[test_params]["STIM_VOLTAGE"],
                                                stimulation_duration=stim_configs[test_params]["STIMULATION_DURATION"],
                                                stimulation_repetitions=stim_configs[test_params]["STIMULATION_REPETITIONS"],
                                                spontaneous_rec_duration=stim_configs[test_params]["SPONTANEOUS_REC_DURATION"],
                                                pause_duration=stim_configs[test_params]["PAUSE_DURATION"],
                                                pulse_frequency=stim_configs[test_params]["PULSE_FREQUENCY"],
                                                inter_electrode_delay=stim_configs[test_params]["DELAY"],
                                                electrode_dict=dict())
    
    recorder_stim.stimulation_recording_from_multiple_networks(no_of_channels=4)
    
    print(f"... done with experiment {test_params}.")

```
