```python
import pandas as pd

# Assuming you already have these DataFrames loaded
# probab_data, speed_data, and firing_data

# Step 1: Add source_speed and target_speed columns
def add_speed_columns(probab_df, speed_df):
    # Create dictionaries for faster lookups
    speed_lookup = {}
    for _, row in speed_df.iterrows():
        key = (row['filename'], row['unit'])
        speed_lookup[key] = row['speed_ms-1']
    
    # Add source_speed column
    probab_df['source_speed'] = probab_df.apply(
        lambda row: speed_lookup.get((row['filename'], row['source_unit_id']), None), 
        axis=1
    )
    
    # Add target_speed column
    probab_df['target_speed'] = probab_df.apply(
        lambda row: speed_lookup.get((row['filename'], row['target_unit_id']), None), 
        axis=1
    )
    
    return probab_df

# Step 2: Add firing rate and ISI metrics columns
def add_firing_metrics(probab_df, firing_df):
    # Create dictionaries for faster lookups
    metrics = ['firing_rate', 'isi_mean', 'isi_median', 'isi_std', 'isi_cv']
    firing_lookup = {}
    
    for _, row in firing_df.iterrows():
        key = (row['filename'], row['unit_id'])
        values = {metric: row[metric] for metric in metrics}
        firing_lookup[key] = values
    
    # Add source metrics
    for metric in metrics:
        source_col = f'source_{metric}'
        probab_df[source_col] = probab_df.apply(
            lambda row: firing_lookup.get((row['filename'], row['source_unit_id']), {}).get(metric, None),
            axis=1
        )
    
    # Add target metrics
    for metric in metrics:
        target_col = f'target_{metric}'
        probab_df[target_col] = probab_df.apply(
            lambda row: firing_lookup.get((row['filename'], row['target_unit_id']), {}).get(metric, None),
            axis=1
        )
    
    return probab_df

# Apply both functions to add all the required columns
probab_data = add_speed_columns(probab_data, speed_data)
probab_data = add_firing_metrics(probab_data, firing_data)

# Now probab_data contains all the additional columns
# Display the first few rows to verify
print(probab_data.head())
```
