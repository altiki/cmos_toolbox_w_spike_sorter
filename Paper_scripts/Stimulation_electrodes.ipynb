```python
import pickle
import pandas as pd
import numpy as np

# Load the pickle file
with open(os.path.join(MAIN_PATH, f'biTE_new/{filename}'), 'rb') as f:
    data = pickle.load(f)

# Check what's actually in the data structure
# Convert to DataFrame if needed
validated_results = pd.DataFrame(data['validated_results']) if isinstance(data['validated_results'], list) else data['validated_results']

# Now filter for 'good' validation
good_results = validated_results[validated_results['validation'] == 'good']

# Create the output dictionary
output_dict = {
    'recording_electrodes': [],
    'stimulus_electrodes': []
}

# Process each row in the filtered dataframe
for _, row in good_results.iterrows():
    # Add stimulus electrodes
    output_dict['stimulus_electrodes'].append([row['source_electrode'], row['target_electrode']])
    
    # Get the recording electrodes from UNIT_TO_EL dictionary
    source_unit = row['source_unit_id']
    target_unit = row['target_unit_id']
    
    # Get the electrodes for each unit and merge them
    source_electrodes = data['UNIT_TO_EL'][source_unit]
    target_electrodes = data['UNIT_TO_EL'][target_unit]
    merged_electrodes = list(set(source_electrodes + target_electrodes))  # Remove duplicates
    
    # Add to recording electrodes
    output_dict['recording_electrodes'].append(merged_electrodes)

# Flatten stimulus_electrodes and remove duplicates
output_dict['stimulus_electrodes'] = list(np.unique([electrode for pair in output_dict['stimulus_electrodes'] for electrode in pair]))

# Flatten recording_electrodes and remove duplicates
# First, create a flat list of all electrodes from all recordings
all_recording_electrodes = [electrode for sublist in output_dict['recording_electrodes'] for electrode in sublist]
# Then remove duplicates
output_dict['recording_electrodes'] = list(np.unique(all_recording_electrodes))

output_filename = f'ID{CHIP_ID}_DIV{DIV}_N{area}_electrodes.pkl'
ELECTRODE_SELECTION_PATH = '/itet-stor/kvulic/neuronies/single_neurons/1_Subprojects/Neurons_As_DNNs/2_Raw_Data/Electrode_Selections/stimulation_dictionaries'

with open(os.path.join(ELECTRODE_SELECTION_PATH, f'{output_filename}'), 'wb') as f:
    pickle.dump(output_dict, f)

print(f"Successfully created {output_filename} with flattened electrode lists")
```
